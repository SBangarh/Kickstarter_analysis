```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#get portion of data
def get_head(dataframe, num_rows):
    if not dataframe.head(num_rows).empty:
        return dataframe.head(num_rows)
    else:
        return "Your dataframe does not exist!"

#get information
def get_info(dataframe):
    if dataframe.empty:
        return "There is an issue with your dataframe!"
    else:
        return dataframe.info()

#get shape
def df_shape(dataframe):
    if dataframe.shape == (0,0):
        return "Your dataframe does not exist!"
    else:
        return dataframe.shape

#decribe dataframe
def describe_df(dataframe, column_name=None):
    if dataframe.empty == False:
        if column_name:
            return dataframe[column_name].describe()
        else:
            return dataframe.describe()
    else:
        return "Your dataframe does not exist!"


#get columns
def get_columns(dataframe):
    list_length = len(dataframe.columns)
    if list_length >= 0:
        return dataframe.columns
    else:
        return "There are no columns in your dataframe!"

#get column data
def column_data(dataframe, column_name: str):
    if column_name in get_columns(dataframe):
        return dataframe[column_name]
    else:
        return "Your dataframe or column does not exist!"

#check nulls
def check_nulls(dataframe):
    num_nulls = dataframe.isnull().sum()
    if dataframe.empty == False:
        return num_nulls
    else:
        return "There are no nulls in your data."

#check value counts
def unique(dataframe):
    return dataframe.nunique()

#check for dups
def check_dups(dataframe):
    num_dups = dataframe.duplicated().sum()
    if dataframe.empty :
        return f"There are {num_dups} duplicates in your dataframe"
    else:
        return f"There are {num_dups} duplicates in your dataframe"
    
#drop dups
def drop_dups(dataframe):
    dataframe = dataframe.drop_duplicates()
    if check_dups(dataframe) == 0:
        return dataframe
    else:
        return "There are still duplicates in your data. Please check your code."

#make box plots
def make_boxplots(dataframe, column_name):
    sns.boxplot(y=dataframe[column_name])
    return plt.show()

#make histogram
def make_hist(dataframe, column_name = None):
    if column_name:
        return plt.hist(dataframe[column_name])
    else:
        return plt.hist(dataframe)

#make heatmap
def make_heatmap(dataframe):
    correlation = dataframe.corr
    return sns.heatmap(correlation)

#make pairplot
def make_pairplot(dataframe):
    return sns.pairplot(dataframe)

#make scatter plot
def make_scatter(dataframe, x_column, y_column, title: str, x_label: str, y_label: str):
    plt.scatter(x=dataframe[x_column], y=dataframe[y_column])
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    return plt.show()
```
